```python
prefix="P3P3"
coarse= prefix+"bcoarse.txt"
medium=prefix+"bmed.txt"
fine = prefix+"bfine.txt"
[x1,y1]=rp.read_plot(coarse)
[x2,y2]=rp.read_plot(medium)
[x3,y3]=rp.read_plot(fine)
```
